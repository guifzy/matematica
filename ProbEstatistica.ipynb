###Permutação com Repetição

- Quantos anagramas podem ser formados com a palavra MATEMATICA

**OBS:** caso tenha repetição usar a formula: Pn = n!/k!, onde k é a quantidade de cada letra repetida

```python
def count_anagrams(palavra):
  # Usar um dicionario para armazenar as ocorrencias de cada letra
  contagem_letras = {} # dicionario

  # Para cada letra da palavra, será verificado se ela já está presente no dicionario,
  # caso ele não esteja, o contador para essa letra é zerado pos só é levado em consideração
  # os números com ocorrencia
  for l in palavra:
    if l not in contagem_letras:
      contagem_letras[l] = 0
    contagem_letras[l] += 1

  n = len(palavra)
  n_anagramas = factorial(n)
  k = 1
  # Multiplica a quantidade de valores armazenados dentro do dicionario
  for count in contagem_letras.values():
    k *= factorial(count)

  n_anagramas //= k

  return n_anagramas

palavra = "MATEMATICA"
num_anagramas = count_anagrams(palavra)
print(f"Numero de anagramas da palavra '{palavra}': {num_anagramas}")

```

# 1- Probabilidade

Probabilidade clássica: usa o numero de eventos sob o conjunto universo

$p(A)$ = $n(A)/n(U)$



##Frequência Relativa ou Empírica / Probabilidade Frequêncial

$f$ = $n/N$


Príncipais carácterísticas:
*   Sua diferença da clássica se da pois ela se propõe a analisar a quantidade de experimentos, por exemplo:

  *   Uma moeda, atravez da probabilidade clássica, tem uma chance de 50% para cara e coroa, porém, ao lançarmos essa moeda 10 vezes, não nescessáriamente elá irá cair cara em 50% das vezes;


*   Quanto maior o total de experimentos, a probabilidade frequêncial se torna igual a clássica:

> $P = lim_{N->∞} = n(A)/n(U)$

*  Em síntese, quando se envolve quantidade de experimentos, é usada a probabilidae frequencial. Quando se trata da probabilidade de um evento em especifico usa-se a probabilidade clássica;





Simulação de eventos eleatórios para o lançamento de uma moeda, sendo cara = 0 e coroa = 1,

EVENTOS:
*   N=10
*   N=100
*   N=1000
*   N=10000


```python
import random

def lancamento_moeda(N):
    n_caras = 0
    n_coroas = 0

    for _ in range(N):
        # Gera um número aleatório entre 0 e 1
        resultado = random.randint(0, 1)

        # Atualiza os contadores
        if resultado == 0:
            n_caras += 1
        else:
            n_coroas += 1

    # Calcula as frequências relativas
    freq_cara = n_caras / N
    freq_coroa = n_coroas / N

    return freq_cara, freq_coroa, n_caras, n_coroas

# Valores de N
N_valores = [25, 100, 1000, 10000]
for N in N_valores:
    freq_cara, freq_coroa, n_cara, n_coroa = lancamento_moeda(N)
    print(f"Para N = {N}:")
    print(f"n cara: {n_cara}")
    print(f"Frequência relativa de cara: {freq_cara:.4f}")
    print(f"n coroa: {n_coroa}")
    print(f"Frequência relativa de coroa: {freq_coroa:.4f}\n")
```

##Distribuição Binomial-*IMPORTANTE*

 É uma extensão da distribuição de Bernoulli. Enquanto a distribuição de Bernoulli modela o resultado de uma única tentativa, a distribuição binomial modela o número de sucessos em n tentativas independentes.

$$p(X = k)=C(n,k)*(p^k)*((1-p)^k-^n)$$

- Em um determinado estado 70% da população é favoravel a um projeto. Dentre 5 pessoas, qual a probabilidade de 3 serem favoraveis ao projeto?

```python
p = 0.7
k = 3
n = 5

prob = comb(n, k) * (p**k) * ((1-p)**(n-k))

print(f"A probabilidade de serem vaforaveis é de {prob*100} %")
```

## Função de probabilidade

>**Função massa de probabilidade( de uma variavel discreta ):** é o mesmo que todos os valores possíveis de $X$,  $0<=p(X=x_0)<=1$, tal que, a soma da probabilidade destes valores seja igual a probabilidade total:

>$∑(x_iϵX)$ $P(X=x_i)=1$

**EXEMPLOS:**

- No lançamento de um dado honesto em que V.A é 0 se se o valor do dado é par e 1 se o valor do dado for impar, a função da probabilidade é a seguinte:

$P(X=x_0)$ = $\frac12 = 0$ e $\frac12 = 1$

Para provar:

Nota-se que cada valor de $X$ é menor que 0 e maior que 1, $0<=p(X=\frac12)<=1$, e a soma de todos os valores de $X$ é igual a $1$:

$∑$$P(X=x_i)=1$

$P(\frac12) + P(\frac12) = 1$

## Medidas de Variabilidade

> Descrevem a extensão em que os dados de um conjunto estão espalhados ou dispersos em torno

### Amplitude



É a diferença entre o maior e o menor valor de um conjunto de dados. É uma medida simples de

### Variância


É a média dos quadrados das diferenças entre cada valor e a média do conjunto de dados. A variância é útil
porque leva em consideração todas as observações do conjunto, mas os seus valores estão em unidades
quadráticas das unidades originais dos dados.

A variância é útil para entender a consistência ou a homogeneidade dos dados.

No entanto, como seus valores estão em unidades quadráticas das unidades originais dos dados, o desvio padrão (a raiz quadrada da
variância) é mais frequentemente usado para expressar a dispersão dos dados em termos das unidades originais.

AMOSTRA:
$$s^2=\frac{∑^{n}_{i=1}(x_i-\overline{x})^2}{n-1}$$

POPULAÇÂO:
$$σ^2=\frac{∑^{n}_{i=1}(x_i-\mu)^2}{n-1}$$

**Desvio Padrão:**

AMOSTRA:
$$s=\sqrt{s^2}$$
POPULAÇÃO:
$$\sigma=\sqrt{σ^2}$$

#8-Intervalo de Confiança

##Estimação

Quando estamos em determinado parâmetro de uma população, extraimos uma amostra e estudamos seus elementos em busca de uma estimar o parâmetro populacional.

Existem dois tipos de estimação: **extimação pontual** e **extimação intervalar**.

**EXTIMAÇÃO PONTUAL**

*   Uma extimação pontual, extima um valor unico de probabilidade para uma amostra.
*   Ela por ser uma valor único, não fornece por si mesma qualquer informação sobre a precisão e confiabilidade da estimativa.

*   Uma alternativa a isso, seria estimar um intervalo entre valores plausíveis.






##Estimativa Intervalar

Essa estimativa sempre é calculada usando um nível de confiança, que se da pela porcentagem de $1-α$, o nível de confiança implica o intervalo de amostras que inclui um valor plausivel.

Um nível de 95% implica que um valor X está contido em algum intervalo de 95% das amostras, e apenas 5% das amostras dariam um resultado errôneo.



*   Se o nível de confiança for alto e o intervalo resultante bastante restrito, nosso conhecimento do valor do parâmetro será razoavelmente preciso.
*   Por outro lado, um intervalo de confiança muito amplo, passa a ideia de que há muita incerteza com relação ao valor do que estamos estimando.



Para estimativa da amostra temos duas opções de cálculo:


*   Média > 30 = tabela z;
*   Média ≤ 30 e/ou variância populacional desconhecida= tabela t-Student.



$$\overline{X}±z*\frac{S}{\sqrt{n}}$$
$$ou$$
$$\overline{X}±t*\frac{S}{\sqrt{n}}$$

Quando a tabela Z for usada, lembrar de dividir a significancia pela metade, $\frac{α}{2}$

- para uma amostra de 50 observações da população com média desconhecida e desvio padrão de 6, seja a  média amostral 20.5 contrua um intervalo de 95% de confiança para média populacional.

```python
# O valor da significancia dividida por 2 é igual a 47.5, na tabela Z 1.96
s = 6
n = 50
media_amostra = 20.5
z = 1.96

resultado = z * (s / np.sqrt(n))
print(f"I.C[{media_amostra - resultado} ; {media_amostra + resultado}]")
```

Considerando-se um nível de 95% de confiança dessa pesquisa amostral, pode-se concluir que a informação do instituto é ou não correta quanto à média divulgada? Considere o desvio padrão amostral, s = 0,53.​

n: {15,80; 16,20; 15,40; 16,10; 14,60; 15,10; 15,40; 15,40}

#9- Teste de Hipótese

Consiste em atribuir um valor hipotético para o parâmetro e, depois, utilizar a informação da amostra para confirmar ou rejeitar esse valor hipotético

$H_0=$ H nulo, representa a hipótese que será comparada;

$H_1=$ H alternativo, representa o restante, ou seja a possibilidade restante;

image.png

**OBS:** Em casos de curva bilateral, o valor equivale a 100%.
Em casos de unilateral, 50%

Lembre-se, quando a curva é bilateral o valor de alfa é $\frac{𝛂}{2}$

*   Caso a variancia populacional for conhecida, $\sigma$, usar:
$$Zcal=\frac{\overline{X}-μ}{\frac{σ}{\sqrt{n}}}$$
*   Caso seja uma amostra e ela seja $n>30$:
$$Zcal=\frac{\overline{X}-μ}{\frac{s}{\sqrt{n}}}$$
*   Caso a amostra seja $n≤30$:
$$Tcal=\frac{\overline{X}-μ}{\frac{s}{\sqrt{n}}}$$



**OBS:** $Zcalc$ e $Tcalc$ são equivalentes a tabela $Z$ e tabela $Student$

$\overline{X}=$ Média da amostra;

$μ=$ Média da população;

**Nível de significância:** Representa o quanto estamos dispostos a aceitar uma porcentagem de erro do Tipo I.

**EXEMPLO:**

- Uma pesquisa afirma que as pessoas gastam no máximo 6,70 h no celular por dia. Para testar essa
afirmação, uma amostra aleatória simples de tamanho 200 forneceu uma média de 7,25 h e desvio
padrão amostral de 2,5 h. Para 1% de significância, verifique tal afirmativa.

```python
# primeira etapa é identificar os parametros
media_pop = 6.7
n = 200
media_amos = 7.25
s = 2.5

# após isso devemos descobrir a area da região de rejeição 1-alpha
# para isso usamos a tabela Z

alpha = (2.32 + 2.33)/2

# com o valor da região de rejeição, basta aplicarmos a formula para
# Descobrir se o H0 está correto

result = (media_amos-media_pop ) / (s/np.sqrt(n))

if result < alpha:
  print("A hipótese nula não foi rejeitada")
else: print("A Hipótese nula foi rejeitada")
```

## Estatística de Teste

Usamos a fórmula da estatística quando testamos hipóteses sobre proporções, dados categóricos (sim/não, sucesso/falha, preferências, etc.);

$Zcalc=\frac{po- p_0}{\sqrt\frac{{p_0(1-p_0)}}{n}}$

$po =$ porporção observada;

$p_0=$ probabilidade do evento;

$p=$ $H_0 e H_1$

## P-Valor-*IMPORTANTE*

O valor-p (p-value) é a probabilidade de observar um resultado estatístico tão extremo quanto, ou mais extremo do que, o observado, sob a hipótese nula. Em outras palavras, o valor-p quantifica a evidência contra a hipótese nula.



*   Rejeita-se $H_0$ se: $p-valor≤𝛼$
*   Não se rejeita se: $p-valor>𝛼$



**EXEMPLO:**

Suponha que estamos testando uma nova droga e queremos verificar se ela é eficaz no tratamento de uma doença. Nossa hipótese nula ($H_0$) seria que a droga não tem efeito, enquanto a hipótese alternativa ($H_1$) seria que a droga tem algum efeito.

Se conduzirmos um teste de hipóteses e obtivermos um valor-p de 0.03 com um nível de significância de 0.05, isso significa que há uma probabilidade de 3% de encontrarmos um resultado tão extremo quanto o observado, se a droga não tiver efeito. Portanto, podemos rejeitar a hipótese nula e concluir que a droga parece ter algum efeito.
